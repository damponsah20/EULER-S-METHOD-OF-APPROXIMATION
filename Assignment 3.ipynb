```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

#Use the Euler's method with step size h= 0.2 to find the approximate values of the solution of the initial value problem
#y= 2x**3 + x**2. Start at (x,y)= (0,1) , 0<= x <= 2. 
#Compare these approximate values with the exact solution



# Define the differential equation dy/dx = 6x^2 + 2x
def f(x, y):
    return 6*x**2 + 2*x

#EULER METHOD 
def euler_method(f, x0, y0, h, x_end):
    steps = int((x_end - x0) / h) #number of steps
    x_values = np.linspace(x0, x_end, steps + 1)
    y_values = np.zeros(steps +1)
    y_values[0] = y0
    for i in range(steps):
        y_values[i+1] = y_values[i] + h * f(x_values[i], y_values[i])
    return x_values, y_values

# Exact solution
def exact_solution(x):
    return 2*x**3 + x**2 

# Initial conditions
x0 = 0
y0 = 1
h = 0.4
x_end = 2

# Compute the approximate values using Euler's method
x_values, y_approx = euler_method(f, x0, y0, h, x_end)

# Compute the exact values for comparison
y_exact = exact_solution(x_values)
y_exact[0]=1



print("Approximate values:")
for x, y in zip(x_values, y_approx):
    print(f"x = {x:.1f}, y_approx = {y:.4f}")

print("\nExact values:")
for x, y in zip(x_values, y_exact):
    print(f"x = {x:.1f}, y_exact = {y:.4f}")


```

FOR VISUALIZATION

```python
plt.figure(figsize=(10,6))
plt.plot( x_values, y_approx, label= " Euler's method", marker ='o')
plt.plot( x_values, y_exact, label= ' Exact_values', marker ='o')
plt.legend()
plt.title("COMPARISON OF EULER'S METHOD AND EXACT VALUES")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, linestyle='--')
```

ERROR CALCULATIONS

```python
def calculate_errors(y_approx, y_exact):
    absolute_errors = [round(abs(y_exact_i - y_approx_i), 3) for y_approx_i, y_exact_i in zip(y_approx, y_exact)]
    relative_errors = [round(abs((y_exact_i - y_approx_i) / y_exact_i) * 100 ,3) for y_approx_i, y_exact_i in zip(y_approx, y_exact)]
    return absolute_errors, relative_errors

absolute_errors, relative_errors = calculate_errors(y_approx, y_exact)

print("Absolute errors:", absolute_errors)
print("Relative errors(%):", relative_errors)
```
